```python
import requests
from bs4 import BeautifulSoup 
```

```python
data_array =[]

def getData(id):
    
    #target url
    url = 'https://www.bbc.com/pashto/topics/cr50y57xj52t='+str(id)
    print(url)
    
    #open the url with get method
    resp = requests.get(url)
    
    #check status of response
    if resp.status_code==200:
        
        #parser, we use default html parser
        soup=BeautifulSoup(resp.content,'html.parser',from_encoding="utf-8")
        
        #get content of the article body 
        full_txt = soup.find('div', {"class": 'content_txt'})
        author = soup.find('div', {"class": 'author_txt'})
        title = soup.find('div', {"class": 'title_txt'})
        dic = {'title':title.text, 'author':author.text, 'content':full_txt.text}
        data_array.append(dic)
        #print(data_array[0]) 
        
    else:
        print("Error.")
            
```

```python
for i in range(30500,32500):
    getData(i)
df = pd.DataFrame(data_array)
df.to_csv('pashto_dataset_uncleaned_30500_32499.csv', encoding='utf-8')
```
